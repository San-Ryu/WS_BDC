<문제>  
세 가지 다른 교육 방법(A, B, C)을 사용하여 수험생들의 시험 성적을 개선시키는 효과를 평가하고자 한다. 30명의 학생들을 무작위로 세 그룹으로 배정하여 교육을 실시하였고, 시험을 보고 성적을 측정하였습니다. 다음은 각 그룹의 학생들의 성적 데이터입니다.  
- 귀무가설(H0): 세 그룹(A, B, C) 간의 평균 성적 차이가 없다.  
- 대립가설(H1 또는 Ha): 세 그룹(A, B, C) 간의 평균 성적 차이가 있다.  
1. 다음 주어진 데이터로 일원배치법을 수행하여 그룹 간의 평균 성적 차이가 있는지 검정하세요  
1) f값 (소수 둘째자리)  
2) p값 (소수 여섯째자리)  
3) 검정결과 출력  

```python
import pandas as pd

## 각 그룹의 데이터
group01 = [85, 92, 78, 88, 83, 90, 76, 84, 92, 87]
group02 = [79, 69, 84, 78, 79, 83, 79, 81, 86, 88]
group03 = [75, 68, 74, 65, 77, 72, 70, 73, 78, 75]

data = {'Group01' : group01, 'Group02' : group02, 'Group03' : group03}
df_wide = pd.DataFrame(data)
df_long = df_wide.melt(value_vars = ['Group01', 'Group02', 'Group03'], var_name = 'Group', value_name = 'Score')
display(df_wide)
display(df_long)
```

```python
from scipy.stats import f_oneway

## 일원배치법
fv, pv = f_oneway(group01, group02, group03)

print(round(fv, 4))
print(format(pv, '.6f'))
```
