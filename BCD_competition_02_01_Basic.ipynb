```python
## 라이브러리 불러오기
import pandas as pd
```

```python
## 데이터 불러오기
train = pd.read_csv("/content/drive/MyDrive/data_BDC/car_market_BDC2022-4th/train.csv")
test = pd.read_csv("/content/drive/MyDrive/data_BDC/car_market_BDC2022-4th/test.csv")

print(train.shape, test.shape)
display(train.head(5), test.head(5))
```

```python
print(train['Segmentation'].value_counts())
print('\n')
print(train.isnull().sum())
print()
print(test.isnull().sum())
```

```python
## 데이터 전처리
## Target값 분리
train_y = train.pop('Segmentation')

## Test 데이터로부터 ID Column 복사
test_ID = test.pop('ID')
```

```python
## 수치형 데이터 Column 분리
## ['ID', 'Age', 'Work_Experience', "Family_size', 'Segmentation"]
list_cols = ['Age', 'Work_Experience', 'Family_Size']

## 수치형 데이터 Column
train_x, test_x = train[list_cols], test[list_cols]
```

```python
## Model 학습 및 예측
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 0)
rf.fit(train_x, train_y)
pred = rf.predict(test_x)
pred
```

```python
## 예측 결과 -> 데이터프레임으로 재생성
## pd.DataFrame({'cust_id' : test_x.cust_id, 'gender' : pred}).to_csv('submission.csv', index = False)

submit = pd.DataFrame({
    'ID' : test_ID
    , 'Segmentation' : pred
})

submit.to_csv('/content/drive/MyDrive/data_BDC/car_market_BDC2022-4th/submission.csv', index = False)
submit
```
