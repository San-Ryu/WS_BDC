```python
import pandas as pd
```

```python
## 데이터 불러오기
path_dir = '/content/drive/MyDrive/data_BDC/car_market_BDC2022-4th/'

train = pd.read_csv(path_dir + 'train.csv')
test = pd.read_csv(path_dir + 'test.csv')
display(train.head(5), test.head(5))
```

```python
## Label Encoding
## Series.astype('category').cat.codes
list_col_train = list(train.describe(include = 'O').columns)
list_col_test = list(test.describe(include = 'O').columns)

for col in list(list_col_train) :
  train[col] = train[col].astype('category').cat.codes

for col in list(list_col_test) :
  test[col] = test[col].astype('category').cat.codes
```

```python
display(train.head(5))
display(test.head(5))
```

```python
## ID값과 Target값 처리
train_x = train.drop('ID', axis = 1)
train_y = train_x.pop('Segmentation')

test_x = test
test_id = test_x.pop('ID')

display(train_x.head(5))
display(train_y.head(5))
display(test_x.head(5))
display(test_id.head(5))
```

```python
## 모델 선택
## 하이퍼파라미터 튜닝 : max_depth, n_estimators
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 0, max_depth = 7, n_estimators = 500)
```

```python
## 교차 검증
from sklearn.model_selection import cross_val_score
scores = cross_val_score(rf, train_x, train_y, cv = 10, scoring = 'f1_macro')

for score in scores :
  print(format(score, ".4f"))
```

```python
## 모델 학습
rf.fit(train_x, train_y)
pred = rf.predict(test_x)
pred
```

```python
## 예측 결과 출력
## pd.DataFrame({'cust_id' : test_x.cust_id, 'Segmentation' : pred}).to_csv('submission.csv', index = False)

submit = pd.DataFrame({
    'ID' : test_id
    , 'Segmentation' : pred
})

submit.to_csv(path_dir + 'submission.csv', index = False)
submit
```
