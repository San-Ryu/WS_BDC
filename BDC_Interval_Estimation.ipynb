- 점추정: 데이터셋을 기반으로 이 도시의 평균 연간 온도를 점추정하세요. (반올림하여 소수 둘째자리까지)  
- 구간추정: 95% 신뢰수준에서 이 도시의 평균 연간 온도에 대한 신뢰구간을 구하세요. (반올림하여 소수 둘째자리까지)

```python
import pandas as pd

temperature_data = pd.read_csv("/content/drive/MyDrive/data_BDC/daily_temperatures.csv")
temperature_data
```

```python
## 점추정 : 샘플 평균 계산
sample_mean = temperature_data['Daily Average Temperature'].mean()
sample_mean
```

```python
from scipy import stats

## 구간추정 : 샘플 표준 편차 및 신뢰구간 계산
confidence_level = 0.95 ## 95% 신뢰수준
sample_std = temperature_data['Daily Average Temperature'].std()
n_samples = len(temperature_data) ## 총 샘플 수 n (자유도는 n-1)

confidence_interval = stats.t.interval(confidence_level, df = n_samples-1, loc=sample_mean, scale=sample_std/(n_samples**0.5))
```
