# 쌍체 T 검정  
주어진 데이터는 고혈압 환자 치료 전후의 혈압이다. 해당 치료가 효과가 있는지 대응(쌍체)표본 t-검정을 진행하시오  
- 귀무가설(H0): μ >= 0  
- 대립가설(H1): μ < 0  
- μ = (치료 후 혈압 - 치료 전 혈압)의 평균  
- 유의수준: 0.05  

1) μ의 표본평균은?(소수 둘째자리까지 반올림)  
2) 검정통계량 값은?(소수 넷째자리까지 반올림)  
3) p-값은?(소수 넷째자리까지 반올림)  
4) 가설검정의 결과는? (유의수준 5%)

```python
import pandas as pd
from scipy import stats

df = pd.read_csv('/content/drive/MyDrive/data_BDC/high_blood_pressure.csv')
# df['diff'] = df['bp_post'] = df['bp_pre']
df['diff'] = df['bp_post'] - df['bp_pre']
df
```

```python
## 01. m의 표본 평균
mean_bp = round(df['diff'].mean(), 2)
print(mean_bp)

## 02, 03. 검정통계량과 p-value
st, pv = stats.ttest_rel(df['bp_post'], df['bp_pre'], alternative = "less")
print(round(st, 4))
print(round(pv, 4))

## 04. 유의수준 0.05에서의 가설검정의 결과는?
## 귀무가설 : 치료 전의 혈압(bp_pre)과 치료 후의 혈압(bp_pose)의 평균에는 차이가 없다.
## 대립가설 : 치료 전의 혈압(bp_pre)과 치료 후의 혈압(bp_pose)의 평균은 같지 않다.
## p-value(0.0015)가 유의수준 0.05보다 작으므로, 귀무가설이 기각되어 대립가설이 채택된다.
## --> 두 집단의 평균은 같지 않다.
```
